```python
import pandas as pd
import seaborn as sns

import matplotlib.pyplot as plt
df = pd.read_csv('heart.csv')
df
```

```python
df[['sex', 'target']].groupby(['sex'], as_index=False).mean().sort_values(by='target', ascending=False)
```

```python
df[['cp', 'target']].groupby(['cp'], as_index=False).mean().sort_values(by='target', ascending=False)
```

```python
def interval(age):
    if 0<=age and age<40:
        return '0-40'
    elif 40<=age and age<50:
        return '40-50'
    elif 50<=age and age<60:
        return '50-60'
    elif 60<=age and age<70:
        return '60-70'
    elif 70<=age:
        return '70-100'
    else:
        return 'none'

grouped=df.set_index('age').groupby(interval)['target'] #利用分组函数可以传递函数对索引列进行分组的特性，先将B设置为索引，再执行分组方法
grouped.count() #调用count方法获取分组数量
```

```python
df[['age', 'target']].groupby(['age'], as_index=False).mean().sort_values(by='target', ascending=False)



```

```python
grid = sns.FacetGrid(df, col='cp', row='sex', height=2.2, aspect=1.6)
grid.map(plt.hist, 'target', alpha=.5, bins=20)
grid.add_legend();
```
